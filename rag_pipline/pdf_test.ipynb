## Chapter 3: Smart Chunking (Preparing for LLMs/RAG) \\

Now that you can extract text and images, the next logical step is preparing that data for an LLM (like OpenAI, Claude, or a local model).
You cannot just dump a 100-page PDF into a prompt. You need to break it into smaller pieces. This is called Chunking.
Most tools (like LangChain's RecursiveCharacterTextSplitter) slice text blindly based on character count. Docling is smarter. Because it understands the document layout (headers, lists, tables), it chunks content semantically.
The HybridChunker
---
Docling provides a HybridChunker. It tries to keep headers with their paragraphs and keeps tables intact, ensuring the AI doesn't lose context.
### Task: Split a document into smart chunks

```python
from docling.document_converter import DocumentConverter
from docling.chunking import HybridChunker

# 1. Convert the document (as usual)
converter = DocumentConverter()
result = converter.convert(file_path)
doc = result.document

# 2. Initialize the Chunker
# tokenizer: We usually specify the model we are using so it counts tokens correctly.
# If you leave it blank, it defaults to a standard tokenizer.
chunker = HybridChunker(
    tokenizer="sentence-transformers/all-MiniLM-L6-v2" 
)

# 3. Generate Chunks
chunk_iter = chunker.chunk(doc)

# 4. Print the results
print(f"Total chunks generated: {len(list(chunker.chunk(doc)))}")

for i, chunk in enumerate(chunk_iter):
    if i < 3: # Just printing the first 3 for brevity
        print(f"\n--- Chunk {i} ---")
        print(f"TEXT: {chunk.text}")
        
        # 5. Access Metadata (Crucial for RAG/Citations)
        # This tells you exactly where in the PDF this text came from
        if chunk.meta.headings:
            print(f"SECTION: {chunk.meta.headings[0]}")
        print(f"PAGE: {chunk.meta.page_numbers}")
```

#### Batch Processing in Python
If you are building an app, you shouldn't loop through files one by one using converter.convert(). That is inefficient.
Instead, use converter.convert_all(). This method is optimized to handle lists of files.

```python
from docling.document_converter import DocumentConverter
from pathlib import Path

# 1. Setup input files
# Let's pretend we have a list of files or URLs
inputs = [
    "https://arxiv.org/pdf/2408.09869",  # File 1
    "https://arxiv.org/pdf/2206.01062",  # File 2
    # Path("./local_file.pdf")           # You can mix Paths and URLs
]

# 2. Initialize Converter
converter = DocumentConverter()

# 3. Use convert_all for batch processing
# This returns an iterator (it processes as you loop)
conv_results = converter.convert_all(inputs)

# 4. Handle results
for result in conv_results:
    if result.errors:
        print(f"Error converting {result.input.file}: {result.errors}")
    else:
        # Get the filename/source name
        filename = result.input.file.name or "document"
        
        # Save to Markdown
        output_md = result.document.export_to_markdown()
        
        print(f"Successfully converted: {filename} ({len(output_md)} characters)")
```

Here is the code to demonstrate that skill. This script manually iterates through the Docling structure to create chunks exactly how you want them.

```python
from docling.document_converter import DocumentConverter

converter = DocumentConverter()
result = converter.convert("https://arxiv.org/pdf/2408.09869")
doc = result.document

# --- CUSTOM CHUNKING LOGIC ---
# Goal: Create chunks where every paragraph includes the metadata of its Parent Section.

custom_chunks = []

# Docling organizes text into 'texts' and tables into 'tables'
# But 'body' contains the ordered list of everything.
for element, level in doc.iterate_items():
    
    # 1. Custom Logic: Skip short headers or page footers
    if isinstance(element, str): # Simple text check (pseudo-code logic)
        if len(element) < 5: continue 
    
    # 2. Identify the type of element
    # Docling gives us labels like 'paragraph', 'heading', 'table', 'caption'
    label = element.label.name 
    
    # 3. Build the chunk text
    text_content = element.text
    
    # 4. ENRICHMENT (The Skill Move):
    # Find the parent section header for this paragraph.
    # (Docling maintains a hierarchy, so you can track the current active header)
    # This is better than LangChain because we know EXACTLY which section we are in.
    
    chunk_entry = {
        "content": text_content,
        "type": label,
        "page": element.prov[0].page_no, # Provenance (Page number)
        "bbox": element.prov[0].bbox.as_tuple() # Coordinates
    }
    
    # 5. Logic: Tables should never be split
    if label == "TABLE":
        # Keep table distinct so we can embed it differently?
        chunk_entry["is_table"] = True
    
    custom_chunks.append(chunk_entry)

print(f"Created {len(custom_chunks)} custom chunks.")
```

Step 1: Binarization (Cleaning Shadows & Noise)
The Problem: A photo of a paper usually has shadows (uneven lighting). OCR engines hate shadows because they look like dark shapes.
The Solution: Adaptive Thresholding. Instead of saying "Every pixel darker than 50% is black," it calculates the threshold for small regions of the image. This removes shadows while keeping text.

```python
import cv2
import numpy as np

def preprocess_binarization(image_path):
    # 1. Read the image
    img = cv2.imread(image_path)
    
    # 2. Convert to Grayscale (OCR doesn't need color)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # 3. Apply Adaptive Thresholding
    # - 255: Maximum value (white)
    # - cv2.ADAPTIVE_THRESH_GAUSSIAN_C: Uses a weighted sum of neighborhood
    # - cv2.THRESH_BINARY: Basic type
    # - 11: Block size (looks at an 11x11 pixel neighborhood)
    # - 2: Constant subtracted from the mean (fine-tuning)
    binary_img = cv2.adaptiveThreshold(
        gray, 255, 
        cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
        cv2.THRESH_BINARY, 
        11, 2
    )
    
    # Optional: Denoising (remove salt-and-pepper noise)
    # perfect for scanned documents
    clean_img = cv2.fastNlMeansDenoising(binary_img, None, 10, 7, 21)
    
    return clean_img

# Usage
clean = preprocess_binarization("dirty_receipt.jpg")
cv2.imwrite("step1_clean.jpg", clean)
```

Step 2: Skew Correction (Deskewing)
The Problem: Users rarely hold their phone perfectly straight. A 5-degree rotation can ruin table extraction.
The Solution:
Detect the block of text.
Draw a "Rotated Rectangle" around it.
Calculate the angle of that rectangle.
Rotate the image back.

```python
def preprocess_deskew(image):
    # 1. Invert colors (White text on black background)
    # We do this because contours are found on white objects
    gray = image # Assuming input is already grayscale/binary from Step 1
    # If input is 0 (black) and 255 (white), invert it:
    bitwise_not = cv2.bitwise_not(gray)

    # 2. Find all non-zero points (text pixels)
    coords = np.column_stack(np.where(bitwise_not > 0))
    
    # 3. Compute the "Minimum Area Rectangle" that covers all text
    angle = cv2.minAreaRect(coords)[-1]
    
    # 4. Handle the angle calculation (OpenCV returns angles strangely)
    if angle < -45:
        angle = -(90 + angle)
    else:
        angle = -angle
        
    print(f"Detected skew angle: {angle} degrees")

    # 5. Rotate the image to correct the skew
    (h, w) = gray.shape[:2]
    center = (w // 2, h // 2)
    
    # Calculate the rotation matrix
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    
    # Perform the actual rotation
    rotated = cv2.warpAffine(
        gray, M, (w, h), 
        flags=cv2.INTER_CUBIC, 
        borderMode=cv2.BORDER_REPLICATE
    )
    
    return rotated
```

The Strategy: The "Text Density" Check
Before you load the heavy Docling engine, use a tiny, fast library like pypdf to peek inside the file.
Try to extract text from the first 1–3 pages.
Count the characters.
If High Text Count (> 50 chars): It is Digital. Turn OCR OFF.
If Low Text Count (< 50 chars): It is a Scan (or an image). Turn OCR ON.

```python
import pypdf

def needs_ocr(pdf_path, check_pages=3, char_threshold=50):
    """
    Returns True if the PDF is likely a scan (needs OCR).
    Returns False if the PDF is likely digital (text layer exists).
    """
    try:
        reader = pypdf.PdfReader(pdf_path)
        
        extracted_text = ""
        
        # Check the first few pages (checking just page 1 can be misleading)
        # We limit to min(check_pages, total_pages) to avoid errors on single-page docs
        num_pages_to_check = min(check_pages, len(reader.pages))
        
        for i in range(num_pages_to_check):
            page_text = reader.pages[i].extract_text()
            if page_text:
                extracted_text += page_text

        # Logic: If we found substantial text, it's digital.
        # We use a threshold because sometimes scans have "artifacts" or 
        # watermarks that look like 1 or 2 characters.
        if len(extracted_text.strip()) > char_threshold:
            return False # It is Digital, NO OCR needed
        else:
            return True  # It is Empty/Image, OCR IS needed

    except Exception as e:
        print(f"Error reading PDF text layer: {e}")
        # If the file is corrupt or unreadable by pypdf, 
        # assume we need the heavy guns (OCR) to try and save it.
        return True
```

```python
# Phase 1: The "Splitter" Script
# We don't just "load" the document. We dissect it. We will use Docling to separate Tables from Text.
```

```python
import pandas as pd
from sqlalchemy import create_engine
from docling.document_converter import DocumentConverter
from langchain_core.documents import Document

# 1. Setup
# We will use SQLite for the tables for simplicity, but use Postgres in production
engine = create_engine("sqlite:///financial_data.db")
converter = DocumentConverter()

def ingest_financial_report(pdf_path):
    print(f"--- Processing {pdf_path} ---")
    result = converter.convert(pdf_path)
    doc = result.document
    
    # --- BUCKET 1: TEXT & IMAGE DESCRIPTIONS ---
    text_docs = []
    
    # Export text content (and image descriptions if VLM is on)
    # We use Docling's chunker or just iterate paragraphs
    for i, text_item in enumerate(doc.texts):
        # Create a LangChain Document
        lc_doc = Document(
            page_content=text_item.text,
            metadata={
                "source": pdf_path,
                "type": "text",
                "page": text_item.prov[0].page_no
            }
        )
        text_docs.append(lc_doc)
        
    # --- BUCKET 2: TABLES (The SQL Part) ---
    table_metadata = []
    
    for i, table in enumerate(doc.tables):
        # 1. Convert to Pandas DataFrame
        df = table.export_to_dataframe()
        
        # 2. Clean the Dataframe (Crucial for SQL)
        # Remove empty rows, ensure column names are strings
        df.columns = [str(c) for c in df.columns]
        df = df.dropna(how='all')
        
        # 3. Create a unique Table Name
        # e.g., "table_page_5_index_0"
        page_no = table.prov[0].page_no
        table_name = f"tab_p{page_no}_{i}"
        
        # 4. Push to SQL Database
        df.to_sql(table_name, engine, if_exists='replace', index=False)
        print(f"Saved Table to SQL: {table_name}")
        
        # 5. Create a 'Summary Document' for the Vector DB
        # We need the Vector DB to know this table EXISTS, so the Agent can find it.
        # We store the *Schema* and a *Summary* in the Vector DB.
        table_summary = f"Financial Table containing columns: {list(df.columns)}. Located on page {page_no}."
        
        table_doc = Document(
            page_content=table_summary,
            metadata={
                "source": pdf_path,
                "type": "table_reference", # <--- Important flag
                "sql_table_name": table_name, # <--- Link to SQL
                "page": page_no
            }
        )
        text_docs.append(table_doc)

    return text_docs

# Run the ingestion
docs_for_vector_db = ingest_financial_report("https://arxiv.org/pdf/2408.09869")
```

Phase 2: Storing in Vector DB
Now push the text_docs (which include paragraphs + table summaries) to your Vector Store.

```python
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Create Vector DB
vector_db = Chroma.from_documents(
    documents=docs_for_vector_db,
    embedding=embeddings,
    collection_name="financial_report"
)

print("Vector DB Populated.")
```

Phase 3: The "Router" (How to know when to call what)
This is the most important part. You asked: "How to know when to call that table data?"
We build an Agent with two tools:
Text Retriever: "Search for text info."
SQL Analyst: "Query specific table data."
When the Vector DB returns a chunk with type: "table_reference", the Agent sees the sql_table_name and switches to the SQL tool.

```python
from langchain.agents import create_tool_calling_agent, AgentExecutor
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI
import pandas as pd

# Initialize LLM (You need a smart model like GPT-4o or Claude 3.5 Sonnet for SQL logic)
llm = ChatOpenAI(model="gpt-4o", temperature=0)

# --- TOOL 1: VECTOR SEARCH (For Text/Concepts) ---
@tool
def search_text_documents(query: str):
    """Useful for finding general information, summaries, and written context."""
    results = vector_db.similarity_search(query, k=3)
    return "\n".join([d.page_content for d in results])

# --- TOOL 2: TABLE ANALYTICS (For Numbers/Math) ---
@tool
def analyze_financial_table(table_name: str, question: str):
    """
    Useful ONLY when you need to do math or look up exact numbers in a specific table.
    Input: the SQL table name and the question.
    """
    # 1. Load the table from SQL back to Pandas
    try:
        df = pd.read_sql_table(table_name, engine)
    except Exception:
        return "Table not found."

    # 2. Use a 'Pandas Agent' logic here (simplified for this example)
    # We dump the dataframe to a string (or CSV) so the LLM can read it.
    # For massive tables, use a SQL query generator instead.
    return f"Data from {table_name}:\n{df.to_markdown()}\n\nQuestion: {question}"

# --- THE AGENT (The Router) ---

tools = [search_text_documents, analyze_financial_table]

from langchain import hub
prompt = hub.pull("hwchase17/openai-functions-agent")

agent = create_tool_calling_agent(llm, tools, prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# --- USAGE EXAMPLES ---

# Scenario A: Text Question
# The Agent will choose 'search_text_documents'
print("--- QUERY 1 ---")
agent_executor.invoke({"input": "What is the main conclusion of the paper?"})

# Scenario B: Table Question
# 1. The user asks a math question.
# 2. The Agent might search Vector DB first.
# 3. Vector DB returns: "Financial Table containing columns: [Revenue, Year]... sql_table_name: tab_p5_0"
# 4. The Agent sees this, and calls 'analyze_financial_table("tab_p5_0", "Calculate year over year growth")'
print("\n--- QUERY 2 ---")
agent_executor.invoke({"input": "Look at the Revenue table on page 5 and calculate the growth from 2023 to 2024."})
```

2. File Hashing (The Cost Saver)
The Risk: User A uploads "Annual_Report_2024.pdf". Ten minutes later, User B uploads the exact same file.
If you run OCR/Docling again, you are burning GPU money and wasting time.
The Solution: Compute the MD5/SHA256 Hash of every file before processing.

```python
import hashlib

def get_file_hash(file_path):
    hasher = hashlib.md5()
    with open(file_path, "rb") as f:
        buf = f.read()
        hasher.update(buf)
    return hasher.hexdigest()

# Logic
file_hash = get_file_hash("upload.pdf")
if check_database_for_hash(file_hash):
    print("File already processed! Retrieving from DB...")
    # Skip Docling, Skip OCR. Instant result.
else:
    process_with_docling(file_path)
```

4. Handling Password-Protected Files
The Problem: Banks and HR departments often password-protect PDFs. Docling will crash if you feed it a locked file.
The Solution: Detect it early (using pypdf again).

```python
from pypdf import PdfReader

def is_locked(file_path):
    try:
        reader = PdfReader(file_path)
        if reader.is_encrypted:
            return True
    except:
        return True # Assume corrupt or locked
    return False

# Usage
if is_locked("bank_stmt.pdf"):
    # ASK USER FOR PASSWORD on the UI
    password = input("Enter PDF Password: ")
    # Docling can accept passwords via pipeline options if needed, 
    # or you unlock it with pypdf first then save a temp unlocked copy.
```
